# Lab 9 — 2023 Yellow Taxi (pandas)

This page contains my Lab 9 analysis using the 2023 NYC Yellow Taxi sample dataset.


```python
import pandas as pd

df = pd.read_csv("yellow_taxi_2023_sample.csv")

print("Columns in dataset:", df.columns.tolist())

df["trip_distance"]
```

```python
average_distance = df["trip_distance"].mean()
print(f"The average ride distance is {average_distance:.2f} miles.")
```

This gives us an idea of the typical length of a yellow taxi trip in the 2023 dataset.

```python
credit_card_trips = len(df[df["payment_type"] == 1])
total_trips = len(df)
percent_credit_card = (credit_card_trips / total_trips) * 100

print(f"{percent_credit_card:.2f}% of trips were paid by credit card.")
```

The column `payment_type` contains numeric codes for payment methods.  
We filter for rows where `payment_type == 1`, which means “Credit Card,”  
and divide that count by the total number of trips to get the percentage.

```python

sample_df = df.sample(frac=0.05, random_state=42)
sample_df.to_csv("yellow_taxi_sample_5pct.csv", index=False)

print("Random sample saved as 'yellow_taxi_sample_5pct.csv'")
```

`sample(frac=0.05)` randomly selects 5% of the rows from the dataset.  
We save this sample to a new CSV file named `yellow_taxi_sample_5pct.csv`.  
The `random_state` ensures results are reproducible.
